```python
import seaborn as sns
```

```python
df = sns.load_dataset('titanic')
df
```

```python
#Age Distribution
fig = sns.histplot(df, x='age')  #histplot()- histogram plot
fig.set_title("Age Distribution Of Titanic Passesngers")
fig.set_xlabel("Age")
fig.set_ylabel("frequency")
```

```python
#survival count based on gender
fig = sns.countplot(df, x='survived', hue='sex')  #hue- for displaying category based distribution
fig.set_title("Survival Count Based On Gender")
fig.set_xlabel("Survived")
fig.set_ylabel("Count")
```

```python
#bar graph
fig = sns.barplot(df, x='sex', y='fare', hue='class')
fig.set_title('Average Fare by passenger Class and Gender')
fig.set_xlabel('Sex')
fig.set_ylabel('Average Fare')
```

```python
fig = sns.countplot(df, x='class',palette='viridis')
fig.set_title("Count of Passenger by Class")
fig.set_xlabel("Paseenger Class")
fig.set_ylabel("Count")
```

```python
fig = sns.countplot(df, x='class',palette='coolwarm')
fig.set_title("Count of Passenger by Class")
fig.set_xlabel("Paseenger Class")
fig.set_ylabel("Count")
```
