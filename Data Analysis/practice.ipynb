- row wise selection 
   - iloc - index based selection 
   - loc - label based selection
- column wise selection 
   - single column selection
       - dictionary like selection
       - object like selection
    - multiple column selection
       - dictionary like selection

```python
df.iloc[2]  #select the third row by index
```

```python
df.iloc[-3]  #select the third last row by default
```

```python
df.iloc[2:6]  #select rows from index 2 to 5(exclusive of 6)-slicing
```

```python
df.iloc[[3,5,9,11]]  #select specific rows from index
```

```python
df.iloc[[0,2,4],[2,3]]  #selecting specified number of rows and columns 
```

```python
df = df.set_index('id')  #set 'id' as the index of the DataFrame
df
```

```python
df.loc[2487200875]
```

```python
df.loc[[7129300520,6600060120]]  #select specific row by index label
```
