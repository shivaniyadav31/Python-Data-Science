```python
df = sns.load_dataset('titanic')
df
```

```python
df.isnull().sum()
```

```python
#here all the null values are replaced by mean
avg_age = round(df['age'].mean())
print(avg_age)
```

```python
df['age'] = df['age'].fillna(avg_age)  #fillna - missing values are replaced
df.isnull().sum()
```

```python
cols_to_drop = ['embark_town']
df = df.drop(columns = cols_to_drop)
df.columns
```

```python
#dropna - to drop missing values
df = df.dropna(subset=['embarked'])
df.isnull().sum()
```
